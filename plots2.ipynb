```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import os
import numpy as np
```

```python
qunatized_512 = pd.read_csv("./results/QUANTIZED_CPU_timings_512_100_1.csv")
qunatized_768 = pd.read_csv("./results/QUANTIZED_CPU_timings_768_100_1.csv")
unquantized_512 = pd.read_csv("./results/UNQUANTIZED_CPU_timings_512_100_1.csv")
unquantized_768 = pd.read_csv("./results/UNQUANTIZED_CPU_timings_768_100_1.csv")
```

```python
print("512 quantized time:", qunatized_512["total_time"].values)
print("512 unquantized time:", unquantized_512["total_time"].values)
print("768 quantized time:", qunatized_768["total_time"].values)
print("768 unquantized time:", unquantized_768["total_time"].values)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(5, 5))
if not isinstance(ax, np.ndarray):
    ax = np.array([ax])
# draw a single bar at a time
ax[0].bar("Regular", unquantized_512["total_time"].values, label="unquantized 512")
ax[0].bar("Quantized", qunatized_512["total_time"].values, label="quantized 512")

ax[0].set_ylabel("Time (s)")
# ax[0].legend(loc="lower left")
ax[0].set_title("512x512 Images")

ax[1].bar("Regular", unquantized_768["total_time"].values, label="quantized 768")
ax[1].bar("Quantized", qunatized_768["total_time"].values, label="quantized 768")

ax[1].set_ylabel("Time (s)")
# ax[1].legend(loc="lower left")
ax[1].set_title("768x768")

fig.tight_layout()
fig.savefig("temp.png")
```

```python
diff = float(qunatized_512["total_time"].values[0]) - float(unquantized_512["total_time"].values[0])
percent = diff / float(unquantized_512["total_time"].values[0]) * 100
print("512 quantized is {:.2f}% faster than unquantized".format(percent))

diff = float(qunatized_768["total_time"].values[0]) - float(unquantized_768["total_time"].values[0])
percent = diff / float(unquantized_768["total_time"].values[0]) * 100
print("768 quantized is {:.2f}% faster than unquantized".format(percent))
```
